```python
# Lemmatizing and slicing out adjectives
def lemmatize(T1970_1): 
    words = ' '.join(T1970_1)
    doc = nlp(words)
    return [token.lemma_ for token in doc]
T1970_1 = lemmatize(T1970_1)
```

```python
d = []
nlp = en_core_web_lg.load()
doc = nlp(" ".join(T1970_1))
for token in doc:
    d.append((token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
          token.shape_, token.is_alpha, token.is_stop))

T1970_pd1 = pd.DataFrame(d, columns=("text", "lemma", "pos", "tag", "dep", "shape", 
                               "is_alpha", "is_stop"))
T1970_pd1.head()
```

```python
# Counting the Frequencies
T1970_freq = T1970["text"].value_counts()
```
