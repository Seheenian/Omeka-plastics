```python
import en_core_web_lg
nlp = en_core_web_lg.load()
```

```python
# Lemmatizing
def lemmatize(T2010): 
    words = ' '.join(T2010)
    doc = nlp(words)
    return [token.lemma_ for token in doc]
T2010 = lemmatize(T2010)
```

```python
d = []
nlp = en_core_web_lg.load()
doc = nlp(" ".join(T2010))
for token in doc:
    d.append((token.text, token.lemma_, token.pos_, token.tag_, token.dep_,
          token.shape_, token.is_alpha, token.is_stop))

T2010_pd = pd.DataFrame(d, columns=("text", "lemma", "pos", "tag", "dep", "shape", 
                               "is_alpha", "is_stop"))
T2010_pd.head()
```

```python
# Removing Stopwords
stop2 = list(np.where(T2010_pd["is_stop"] == False)[0])
T2010 = T2010_pd.iloc[stop2]
```

```python
# Counting the Frequencies
T2010_freq = T2010["text"].value_counts()
```

```python
# Transforming to Dictionary and Removing Additional Punctuations and Type errors
T2010_Freq = {k:v for k, v in T2010_freq.iteritems()}
try: 
    del T2010_Freq["-PRON-"]
    del T2010_Freq["·"]
    del T2010_Freq["iso"]
    del T2010_Freq["al"]
    del T2010_Freq["bs"]
    del T2010_Freq["din"]
    del T2010_Freq["et"]
    del T2010_Freq["s"]
    del T2010_Freq['"']
    del T2010_Freq["’"]
    del T2010_Freq["'"]
    del T2010_Freq["◦"]
    del T2010_Freq["−"]
except KeyError: 
    print("key item not found")
```

```python
# Wordcloud Visualization
wordcloud = WordCloud(max_font_size=60, max_words=100, background_color="black", colormap = "PuBuGn_r", random_state=None, ranks_only=False).generate_from_frequencies(T2010_Freq)
plt.figure(figsize = (18, 18))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
wordcloud.to_file("drive/My Drive/DIGHUM/plastic2010_wordcloud.png")
```
